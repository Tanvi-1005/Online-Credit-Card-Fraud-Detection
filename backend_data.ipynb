```python
import pandas as pd
import json

# Load dataset
df = pd.read_csv("final_dataset_11k.csv")

df.head()
```

```python
# Convert to datetime
df["trans_date_trans_time"] = pd.to_datetime(df["trans_date_trans_time"])

# Extract hour (0–23)
df["trans_hour"] = df["trans_date_trans_time"].dt.hour

# Extract day of week (Monday=0 → Sunday=6)
# Convert to 1–7 format (Monday=1, Sunday=7)
df["day_of_week"] = df["trans_date_trans_time"].dt.dayofweek + 1

df[["trans_hour", "day_of_week"]].head()
```

```python
required_columns = [
    "merchant",
    "category",
    "amt",
    "city",
    "state",
    "city_pop",
    "trans_hour",
    "day_of_week",
    "is_fraud"
]

df_final = df[required_columns]
```

```python
df_sample = df_final.sample(n=1000, random_state=42).reset_index(drop=True)

df_sample.head()
```

```python
df_with_label = df_sample.copy()

df_with_label.to_json("fraud_test_data_with_label.json", orient="records", indent=4)
```

```python
df_without_label = df_sample.drop(columns=["is_fraud"])

df_without_label.to_json("fraud_test_data_without_label.json", orient="records", indent=4)
```

```python
print(json.dumps(df_without_label.iloc[0].to_dict(), indent=4))
```

```python
df_with_label.to_csv("fraud_test_data_with_label.csv", index=False)
df_without_label.to_csv("fraud_test_data_without_label.csv", index=False)
```
